```python
# =========================================
# IMPORT LIBRARIES
# =========================================

import pandas as pd
import numpy as np

# =========================================
# LOAD DATA
# =========================================

train_df = pd.read_csv(
    "../data/processed/cleaned_train.csv"
)

test_df = pd.read_csv(
    "../data/processed/cleaned_test.csv"
)
# =========================================
# TOTAL SQUARE FEET
# =========================================

train_df["TotalSF"] = (
    train_df["TotalBsmtSF"] +
    train_df["1stFlrSF"] +
    train_df["2ndFlrSF"]
)

test_df["TotalSF"] = (
    test_df["TotalBsmtSF"] +
    test_df["1stFlrSF"] +
    test_df["2ndFlrSF"]
)
# =========================================
# HOUSE AGE
# =========================================

train_df["HouseAge"] = (
    train_df["YrSold"] -
    train_df["YearBuilt"]
)

test_df["HouseAge"] = (
    test_df["YrSold"] -
    test_df["YearBuilt"]
)
# =========================================
# TOTAL BATHROOMS
# =========================================

train_df["TotalBath"] = (
    train_df["FullBath"] +
    (0.5 * train_df["HalfBath"])
)

test_df["TotalBath"] = (
    test_df["FullBath"] +
    (0.5 * test_df["HalfBath"])
)
# =========================================
# HAS GARAGE
# =========================================

train_df["HasGarage"] = (
    train_df["GarageArea"] > 0
).astype(int)

test_df["HasGarage"] = (
    test_df["GarageArea"] > 0
).astype(int)
# =========================================
# SAVE FEATURED DATA
# =========================================

train_df.to_csv(
    "../data/processed/featured_train.csv",
    index=False
)

test_df.to_csv(
    "../data/processed/featured_test.csv",
    index=False
)

print("Feature engineered datasets saved.")
```
